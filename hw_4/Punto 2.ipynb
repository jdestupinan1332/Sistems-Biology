#### c) Para dos reacciones independientes (A, B) con tasas promedio constantes de r=10 y k=5, genere una serie de tiempo donde ocurran al menos 50 eventos de cada tipo. Haga el histograma de los tiempos entre eventos del mismo tipo y entre eventos en general. Ajuste a cada histograma una exponencial y encuentre sus medias.

```python
def reaction(tasa):
    lista1 =[]
    lista2 =[]
    tau = 0
    u = random.uniform(0,1,50)
    i= 0
    while i < 50:
        ran = u[i]
        tau = tau + log(1/ran)/tasa
        tu2 = log(1/ran)/tasa
        lista1.append(tau)
        lista2.append(tu2)
        i= i+1
    return lista1, lista2
```

```python
x = arange(0,50,1)
rx1, tempos1 = reaction(10)
rx2, tempos2 = reaction(5)
```

```python
plot(rx1,x)
plot(rx2,x)
xlabel('tiempo')
ylabel('reacciones')
legend(['r=10', 'k=5'])
```

#### D) Genere una serie de tiempo para una reacción A con tasa promedio constantes de r=15, pero para cada evento genere un numero aleatorio con probabilidad 1/3 de cambiar el tipo de reacción a B. Haga el histograma de los tiempos entre eventos del mismo tipo y entre eventos en general. Ajuste a cada histograma una exponencial y encuentre sus medias.

```python
def guill (tasa, A0, tau, tb, ta):
    cantA = A0
    u1 = random.uniform()
    u2 = random.uniform()
    tau = tau + log(1/u1)/(cantA*tasa)
    tg = log(1/u1)/(cantA*tasa)
    tb = tb + tg
    ta = - tg + tb
    if u2 < 1/3.0 and cantA > 1:
        cantA = cantA-1
        tb = 0
    
    return cantA, tau, tg, tb, ta
    
```

```python
lista = []
tiempos = []
tG = []
tB = []
tA = []
t=0
tb = 0
ta = 0
A=200
for i in range(500):
    A, t, tg, tb, ta= guill(15, A, t, tb, ta)
    lista.append(A)
    tiempos.append(t)
    tG.append(tg)
    tB.append(tb)
    tA.append(ta)
```
