$$\tau = \frac{1}{r} ln\left(\frac{1}{u}\right)$$

donde u es un numero aleatorio uniformemente distribuido entre 0 y 1, produce una distribución exponencial.

```python
%pylab inline
```

```python
r = arange(0.3,1.8, 0.3) #lista de 0.3 < r < 1.5  medias 
x = arange(0,5,0.01) #lista de vlores de x para poder graficar bien
for media in r:
    F = 1 - pow(e,-media*x) #por cada valor en r se hace esto
    plot (x,F)
xlabel('x')
ylabel('F(x)')
legend (r)
```
